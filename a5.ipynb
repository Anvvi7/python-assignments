```python
initials = "AM"
X = sum(ord(char) for char in initials)

sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales data:", sales)

tax_rate=((X%5)+5)/100
taxed_sales=sales*(1-tax_rate)
print('sales after tax',taxed_sales)

discounted_sales=np.where(sales<X+100,sales*0.95,sales*0.90)
print('sales after discount',discounted_sales)

weekly_sales=np.vstack([discounted_sales]*3)
weekly_sales*=(1+np.arange(3).reshape(3,1)*0.02)
print('weekly sales after 2% increase per week',weekly_sales)
```
